```python
import numpy as np
import matplotlib.pyplot as plt
import random
from mpl_toolkits.mplot3d import Axes3D
import csv
M=[]
M= np.genfromtxt("da.csv", delimiter=",", skip_header=1)
c = np.array([[0.00] * 4] *150)
v = np.array([[0.0000]*4]*3)

for i in range(150):
    for j in range(4):
        c[i][j]=M[i][j]
print("reduced data set with sepal length/width, petal length/width")
print(c)   
c=c.transpose()
Mcov=np.cov(c)
print("\ncovariance of given matrix: \n",Mcov,"\n")
E,V=np.linalg.eig(Mcov)#E=eigen value,V=eigen vector
print("eigen values are: \n",E,"\n")
print("eigen vectors are: \n",V,"\n")

for i in range(3):
    for j in range(4):
        v[i][j]=V[i][j]
print("three vectors chosen for PCA:\n",v,"\n")
M2=np.matmul(v,c)
M2=M2.transpose()
print(M2)

s = []
t = []
u = []
for i in range(50):
    s.append(M2[i][0])
    t.append(M2[i][1])
    u.append(M2[i][2])
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(s, t, u)

x = []
y = []
z = []
for i in range(50):
    x.append(M2[i+50][0])
    y.append(M2[i+50][1])
    z.append(M2[i+50][2])
ax.scatter(x,y,z)

a = []
b = []
c = []

for i in range(50):
    a.append(M2[i+100][0])
    b.append(M2[i+100][1])
    c.append(M2[i+100][2])

ax.scatter(a,b,c)
plt.show()
```
