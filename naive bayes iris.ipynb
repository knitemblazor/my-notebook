```python
import numpy as np
import matplotlib.pyplot as plt
import math
from numpy import *

M=[]
M= np.genfromtxt("da.csv", delimiter=",", skip_header=1)
c = np.array([[0.00] * 4] *150)
v = np.array([[0.0000]*4]*3)

for i in range(150):
    for j in range(4):
        c[i][j]=M[i][j]
#print("reduced data set with sepal length/width, petal length/width")
#print(c)   
c=c.transpose()
Mcov=np.cov(c)
#print("\ncovariance of given matrix: \n",Mcov,"\n")
E,V=np.linalg.eig(Mcov)#E=eigen value,V=eigen vector
#print("eigen values are: \n",E,"\n")
#print("eigen vectors are: \n",V,"\n")
V=V.transpose()
for i in range(1):
    for j in range(4):
        v[i][j]=V[i][j]
#print("one vector chosen for PCA:\n",v,"\n")
M2=np.matmul(v,c)
M2=M2.transpose()
#print(M2)

X1=M2[:,0]


e=[]
for i in range(0,50):
    e.append(0)

#plt.scatter(X1[0:50],e)
#plt.scatter(X1[50:100],e)
#plt.scatter(X1[100:150],e)
#print("\n\nblue = Iris-setosa,orange=Iris-versicolor,green=Iris-virginica\n\n")
#plt.show()

x=X1[0:150]
Q=[]
for i in range(150):
    Q.append(1)
    if(i<50):
        Q[i]=1
    elif(50<=i<100):
        Q[i]=2
    elif(100<=i<150):
        Q[i]=3
#print(Q)

d=[X1[0:150]]
d=np.array(d)     #only one characteristics remain
d=d.transpose()

dtrain=[d[0:35],d[50:85],d[100:135]]
dtest=[d[35:50],d[85:100],d[135:150]]
dtrain=np.array(dtrain)
dtest=np.array(dtest)
dtrain=dtrain.ravel()
dtest=dtest.ravel()
print(len(dtest))
x=X1[0:150]
#print(x,n)
n=35
def mean(x):
    mean=sum(x)/n
    return(mean)
mean1=mean(x[0:35])
mean2=mean(x[50:85])
mean3=mean(x[100:135])
meanx=[mean1,mean2,mean3]
meanx=np.array(meanx)

def std(x,mean):
    std=(sum((x-mean)**2))/n
    std=np.sqrt(std)
    return(std)
std1=std(x[0:35],meanx[0])
std2=std(x[50:85],meanx[1])
std3=std(x[100:135],meanx[2])
stdx=[std1,std2,std3]
stdx=np.array(stdx)
#print(std)
y = []
for i in range(0,3):
    y.append(0)
#train data set 
y[0]=np.exp(-((x[0:35]-meanx[0])**2)/(2*stdx[0]**2))/(stdx[0]*np.sqrt(2*np.pi))
y[1]=np.exp(-((x[50:85]-meanx[1])**2)/(2*stdx[1]**2))/(stdx[1]*np.sqrt(2*np.pi))    
y[2]=np.exp(-((x[100:135]-meanx[2])**2)/(2*stdx[2]**2))/(stdx[2]*np.sqrt(2*np.pi))
a=max(y[0])
b=max(y[1])
c=max(y[2])


z=[a,b,c]
print(z)
#test data set
ytest = np.array([[0.000]*45]*3)

for i in range(3):
    for j in range(45):
        ytest[i][j]=(np.exp(-((dtest[j]-meanx[i])**2)/(2*stdx[i]**2))/(z[i]*stdx[i]*np.sqrt(2*np.pi)))
        ytest[i][j]=round(ytest[i][j],2)

g=[]
for j in range(45):
    g.append(0)
ytest1 = np.array([[0]*45]*3)
for i in range(3):
    for j in range(45):
        if (i==0 and ytest[i][j]>0.2 ):
            g[j]=1
        elif (i==1 and ytest[i][j]>0.5):
            g[j]=2
        elif (i==2 and ytest[i][j]>0.5):
            g[j]=3
        
h=g[0:15]
k=g[15:30]
l=g[30:45]
m=0
r=0
n=0
for i in range(15):
    if(h[i]==1):
        r+=1
    if(k[i]==2):
        m+=1
    if(l[i]==3):
        n+=1
# print(h,k,l)
# print(m+r+n)
print("accuracy:",(m+r+n)/45)
#print(ytest)
# print(g)            
# plt.scatter(x[0:35],y1)
# plt.scatter(x[50:85],y2)
# plt.scatter(x[100:135],y3)
# plt.show()
```
